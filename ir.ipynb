#### first part is importing the libraries. we are using hazm library that is a persian text manipulating library based on NLTK library.

```python
from __future__ import unicode_literals
from hazm import *
import operator
import os
from collections import OrderedDict 
import pickle
import re
import math
```

#### in this part we define normalizer , stemmer and lemmatizer to be used in next part.

#### we also define the structure of dictionary. 

#### the structure would be dictionary = { 'term' : [ total_termfreq , { 'doc' : [ termfreq_in doc , [ 1 , 2 , 3 , 4 ] ] } ] }

```python
normalizer = Normalizer()
stemmer = Stemmer()
lemmatizer = Lemmatizer()
dictionary = {'term':[0,{'doc':[4,[1,2,3,4]]}]}
docs_count=0
docs_size={}
```

#### loading files from a directory and generating dictionary (postings list):

```python
for i in range(3,8):
    path = 'Documents/HamshahriData/HamshahriCorpus/200'+str(i) #enter the directory here
    c=0
    for filename in os.listdir(path):
        docs_count+=1
        f = open(path+'/'+filename, "r")
        text=f.read()
        text=normalizer.normalize(text)
        tokens=word_tokenize(text)
        size=len(tokens)
        docs_size[filename]=size
        position_counter=0
        for t in tokens:
            position_counter+=1
            x=lemmatizer.lemmatize(stemmer.stem(t))
            if x in dictionary:
                dictionary[x][0]=dictionary[x][0]+1
                if filename in dictionary[x][1]:
                    dictionary[x][1][filename][0]=dictionary[x][1][filename][0]+1
                    dictionary[x][1][filename][1].append(position_counter)
                else:
                    dictionary[x][1][filename]=[1,[position_counter]]
            else:
                dictionary[x]=[1,{filename:[1,[position_counter]]}]

```

#### making a list of terms ans frequencies so that we can sort and find out the most used terms, because in our dictionary it's not possible to sort by the term frequencies:

```python
sorted_dictionary=[]
for a in dictionary:
    sorted_dictionary.append((a,dictionary[a][0]))
sorted_dictionary=sorted(sorted_dictionary, key=lambda tup: tup[1])
```

#### in this part we get a query and find the matches using the specified scoring algorithm:

#### first we will tokenize the query and get ready to use the algorithms:

```python
print("please enter your query : ",end="")
query=input()
query_tokens=word_tokenize(query)
m=[]
d=[]
modified_tokens=[]
for a in query_tokens:
    if '*' in a:
        d.append(a)
        a=a+'$'
        a= a.replace("*", "\w+")
        reg= re.compile(a)
        m = m+[string for string in dictionary if re.match(reg, string)]
query_tokens+=m
for v in d:
    query_tokens.remove(v)
    
for o in query_tokens:
    modified_tokens.append(lemmatizer.lemmatize(stemmer.stem(o)))
mod=modified_tokens
modified_tokens=list(set(modified_tokens))
for a in modified_tokens:
    if a not in dictionary:
        modified_tokens.remove(a)
```

## TF-IDF mode

#### making the TF document vectors: (we should use log for tf)

```python
weights={}
for word in dictionary:
    if word in modified_tokens:
        for document in dictionary[word][1]:
            if document not in weights:
                weights[document]=[0]*len(modified_tokens) #making the empty vector for documents 
        index=modified_tokens.index(word)
        for document in dictionary[word][1]:
            weights[document][index]=math.log10(dictionary[word][1][document][0]/docs_size[document])
weights['query']=[0]*len(modified_tokens)
for word in modified_tokens:
    index=modified_tokens.index(word)
    count=mod.count(word)
    weights['query'][index]=math.log10(count/len(mod))
    # and now idf for query!
for i in range(0,len(modified_tokens)):
    weights['query'][i]=weights['query'][i]*(math.log10(docs_count/len(dictionary[modified_tokens[i]][1])))
```

#### in this part we do the cosine normalization if its needed:

```python
print("please choose the scoring mode number: 1. lnn-ltn  2. lnc-ltc ")
mode=input()
if mode=='2':
    for document in weights:
        x=0
        for term in weights[document]:
            x+=term*term
        cosin=math.sqrt(x)
        for i in range(0,len(weights[document])):
            weights[document][i]*=cosin
```

#### now we compute the final scores:

```python
scores=[]
for doc in weights:
    if doc=='query':
        continue
    x=0
    for i in range(0,len(weights[doc])):
        x+=weights[doc][i]*weights['query'][i]
    scores.append((doc,x))
scores.sort(key=lambda tup:tup[1])
scores.reverse()
```

## phrasal search

#### if there is duplicate words in query it will be so difficult! so we just consider the simple form :))

#### the phrasal query doesn't include wildcards and the order is important and also its terms are not repeated so we just use mod instead of modified_tokens:

```python
list_of_docs=[]
for tok in mod:
    li=[]
    if tok not in dictionary:
        print(tok+" is not availabe in dictionary")
        break
    for document in dictionary[tok][1]:
        li.append(document)
    list_of_docs.append(li)
```

#### now we will intersect the sublists to find out the documents that include all the terms:

```python
len(list_of_docs)
```

```python
final_docs_list=list_of_docs[0]
for a in range(0,len(list_of_docs)):
    final_docs_list=list(set(final_docs_list) & set(list_of_docs[a])) 
```

```python
final_docs_list
```

#### now we find documents that the order of the terms ar correct:

```python
result=[]
for doci in final_docs_list:
    terms_positions_list=[]
    for a in mod:
        terms_positions_list.append(dictionary[a][1][doci][1])
    #print(terms_positions_list)
    if len(terms_positions_list)!=len(mod):
        continue
    x=len(terms_positions_list)
    for b in terms_positions_list:
        for i in range(0,len(b)):
            b[i]-=len(terms_positions_list)-x
        x-=1
    mo=terms_positions_list[0]
    for a in range(0,len(terms_positions_list)):
        mo=list(set(mo) & set(terms_positions_list[a])) 
    if len(mo)>0:
        result.append(doci)
```

#### now we compute the precision and recall one by one so that we can compute f measure and map

```python
precision=[]
recall=[]
all_R=len(query_docs[int(nu)-1])
tp=0
retrieved=0
R_possisions=[]
for a in scores[:len(query_docs[int(nu)-1])]:
#for a in scores:
    if a[0] in  query_docs[int(nu)-1]:
        tp+=1
        retrieved+=1
        R_possisions.append(retrieved-1)
        precision.append(tp/retrieved)
        recall.append(tp/all_R)
    else:
        retrieved+=1
        precision.append(tp/retrieved)
        recall.append(tp/all_R)
print("precision is : ",end='')
print(precision[-1])
print("recall is : ",end='')
print(recall[-1])
```

#### and computing f measure and map

```python
beta=1
P=precision[-1]
R=recall[-1]
Fmeasure = (((beta*beta)+1)*P*R)/(beta*beta)*P+R
print("F1 is : ",end='')
print(Fmeasure)
MAP=0
for i in R_possisions:
    MAP+=precision[i]
MAP/=len(R_possisions)
print('MAP is : ',end='')
print(MAP)
```
